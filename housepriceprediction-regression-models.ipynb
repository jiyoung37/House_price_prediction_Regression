# Data preprocessing
To ensure that:

- The dataset contains no missing values before training.

- Numerical variables maintain their original distribution.

- Categorical variables are ready for use in regression models.

## Identifying missing values

```python
#missing data
df.isna().sum()
```

```python
# Make a Function to calculate the percent of missing data in each columns (feature) and then sort it
def missing_percent(df):
    na_percent=100*(df.isna().sum()/len(df))
    na_percent= na_percent[na_percent>0].sort_values()
    return na_percent

na_percent=missing_percent(df)
na_percent
```

## Removing columns with too many missing values
Columns with more than 80% missing values are considered uninformative and are dropped.

```python
# Drop columns with more than 80 % missing values
missing_values=df.isnull().mean()
df=df.loc[:, missing_values<0.8]
```

```python
na_percent=missing_percent(df)
na_percent
```

## Filling missing values in numerical variables
For the remaining numerical features, missing values were imputed using the mean of each variable.

```python
numerical_var = df.select_dtypes(include = ['float64','int64']).columns
for col in numerical_var:
    df[col] = df[col].fillna(df[col].mean())
```

## Encoding categorical variables
Categorical variables are one-hot encoded to be transformed into binary variably. This automatically handles missing values by assigning them to a separate binary column.  

```python
categorical_var = df.select_dtypes(include ='object').columns
df = pd.get_dummies(df, columns = categorical_var)
```

```python
na_percent=missing_percent(df)
na_percent
```

# Modeling preparation

## Feature and Target variable selection

```python
y = df["SalePrice"]
X = df.drop("SalePrice", axis=1)
```

## Train-Test split
Split the dataset into a training set and a test set using 70/30 ratio to evaluate model performance on unseen data. A fixed random_state ensures reproducibility.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Feature standarization
Use StandardScaler to standardize the features since most regression models perform better when features are on the same scale.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Modeling
This section covers the application and evaluation of various machine learning models to predict house prices.
To assess model performance, two key metrics are used:

- R² Score: Measures how well the model explains the variance in the target variable.

- RMSE (Root Mean Squared Error): Measures the average magnitude of the prediction error.

## 1. Linear regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, y_train)
```

```python
# Model evaluation: R2 score
train_score=lin_reg.score(X_train_scaled, y_train)
test_score=lin_reg.score(X_test_scaled, y_test)

print('train score:', train_score)
print('test score:', test_score)

# Model evaluation: RMSE
from sklearn.metrics import mean_squared_error

train_rmse=np.sqrt(mean_squared_error(y_train, lin_reg.predict(X_train_scaled)))
test_rmse=np.sqrt(mean_squared_error(y_test, lin_reg.predict(X_test_scaled)))
print('train RMSE:', train_rmse)
print('test RMSE:', test_rmse)
```
